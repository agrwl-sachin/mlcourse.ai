__2. What was the percentage of male gymnasts among all the male participants of the 2000 Olympics? Round the answer to the first decimal.__

*Hint:* here and further if needed drop duplicated sportsmen to count only unique ones. 

- 0.2
- 1.5 -> Correct
- 2.5
- 7.7 

```python
grouped = data.groupby(['ID', 'Name', 'Sex', 'Age', 'Height', 'Weight', 'Team', 'NOC', 'Games',
       'Year', 'Season', 'City', 'Sport', 'Event', 'Medal'])
index = [gp_keys[0] for gp_keys in grouped.groups.values()]
unique_df = data.reindex(index)
unique_df.count()
```

```python
def find_unique(df):
    df_grouped = df.groupby(['ID'])
    index = [gp_keys[0] for gp_keys in df_grouped.groups.values()]
    unique_data = df.reindex(index)
    return unique_data
```

```python
total_male = (data.loc[(data['Year'] == 2000) & (data['Sex'] == 'M')])
total_male_gymnast = (data.loc[(data['Year'] == 2000) & (data['Sex'] == 'M') & (data['Sport'] == 'Gymnastics')])

unique_total_male = find_unique(total_male)
unique_total_male_gymnast = find_unique(total_male_gymnast)

total_male_cnt = total_male['ID'].count()
total_male_gymnast_cnt = total_male_gymnast['ID'].count()

unique_total_male_cnt = unique_total_male['ID'].count()
unique_total_male_gymnast_cnt = unique_total_male_gymnast['ID'].count()


print( total_male_gymnast_cnt / total_male_cnt )
print( unique_total_male_gymnast_cnt / unique_total_male_cnt )
```
